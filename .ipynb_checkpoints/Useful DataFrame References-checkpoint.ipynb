```python
import pandas as pd
import numpy as np
```

```python
#Random generation of values
samples_list = []
for i in range(0, 20):
    samples_list.append(np.random.random_integers(0,99, 3))
sample_df = pd.DataFrame(samples_list, columns=['A','B','C'])
```

```python
# create new attribute of DataFrame and set it's value to the previous index value
sample_df['last_value_A'] = sample_df['A'].shift(1)
# create new attribute  of the DataFrame and set it's value to a rolling 5 value average
sample_df['rolling_5'] = pd.rolling_mean(sample_df['B'], 5)
sample_df
```
