```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
X_reducded = pca.transform(X)
```

```python
plt.scatter(X_reducded[:, 0], X_reducded[:,1], c=y, cmap='RdYlBu' )
```

# Kmeans Clusters

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=3, random_state=0)
k_means.fit(X)
y_pred = k_means.predict(X)
plt.scatter(X_reducded[:, 0], X_reducded[:,1], c=y_pred, cmap='RdYlBu')
```
